```python
# If we wanted to look at the records for all for all of the cars manufactured
# after 2015, 
df[
    df['Year'] >= 2015 # This is the condition. The nested condition generates a
   # boolean results for all the true values. 
   ]
   # The outer fragment of code displays the condition.
```

```python
df[
    df['Make'] == 'Nissan' # A different example of filtering, but in the same
   # spirit; this nested condition searches for all the Nissan cars
   ]
   # The outer fragment of code displays the condition.
```

```python
# We can combine conditions in the same way:
df[
    (df['Make'] == 'Nissan') & (df['Year'] >= 2015)
   ]
```

```python
# Grouping
```

```python
# In SQL, this query is saying we want to compute the average transmission price
# for the manual transmissions and the automatic transmissions.
# This will come from the cars dataframe, and it will be grouped by
# transmission type in ascending order.
# SELECT
#   transmission_type,
#   AVG(MSRP)
# FROM
#   cars
# GROUP BY
#   transmission_type
```

```python
# This is the SQL query in python:
df.groupby('Transmission Type').MSRP.mean()
```

```python
df.groupby('Transmission Type').MSRP.max()
```

```python
df.groupby('Transmission Type').MSRP.min()
```
